```python
import matplotlib.pyplot as plt

def draw_triangle(ax, p1, p2, p3):
    ax.fill([p1[0], p2[0], p3[0]], [p1[1], p2[1], p3[1]], 'k')

def get_mid(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)

def sierpinski(ax, p1, p2, p3, degree):
    if degree == 0:
        draw_triangle(ax, p1, p2, p3)
    else:
        p12 = get_mid(p1, p2)
        p23 = get_mid(p2, p3)
        p31 = get_mid(p3, p1)
        
        sierpinski(ax, p1, p12, p31, degree-1)
        sierpinski(ax, p12, p2, p23, degree-1)
        sierpinski(ax, p31, p23, p3, degree-1)
        
# Змінна n визначає рівень рекурсії
n = 3

def main():
    p1 = [0, 0]
    p2 = [1, 0]
    p3 = [0.5, 0.866]
    fig, ax = plt.subplots()
    sierpinski(ax, p1, p2, p3, n)
    ax.set_aspect('equal', 'box')
    ax.axis('off')
    plt.show()

main()
```

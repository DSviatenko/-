```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Завантаження файлу
df = pd.read_csv('shopping_trends.csv')

# Розмір таблиці
print("Розмір таблиці:", df.shape)

# Перевірка наявності пропусків
print("\nПропуски в даних:")
print(df.isnull().sum())

# Перевірка наявності дублікатів
print("\nКількість дублікатів:", df.duplicated().sum())

# Типи стовпців
print("\nТипи стовпців:")
print(df.dtypes)

# Перейменування стовпців
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Виведення першіх рядків для перевірки результату
print(df.head())
```

```python
# 1. Гістограма віку (age):
# Побудуйте гістограму віку клієнтів, використовуючи Seaborn.
sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=20, color='skyblue')
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
#  2. Порівняння розподілу покупок за статтею (item_purchased):
#  Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які
#  товари покупають клієнти найчастіше.

plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='item_purchased', palette='viridis', order=df['item_purchased'].value_counts().index)
plt.title('Порівняння розподілу покупок за статтею')
plt.xlabel('Кількість покупок')
plt.ylabel('Товар')
plt.show()
```

```python
#  3. Залежність між віком та сумою покупок (purchase_amount):
#  Використовуйте scatter plot, щоб візуалізувати залежність між віком та
#  сумою покупок.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)', color='skyblue')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.show()
```

```python
#  4. Графік розподілу оцінок товарів (review_rating):
#  Побудуйте графік розподілу оцінок товарів для оцінки якості

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='review_rating', palette='muted')
plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка')
plt.xticks(rotation=45)
plt.ylabel('Кількість')
plt.show()
```

```python
#  5. Категоріальна діаграма для статі (gender):
#  Створіть категоріальну діаграму, щоб відобразити розподіл покупок за
#  статтею.

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', palette='pastel')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#  6. Залежність між рейтингом і сумою покупок (review_rating і
#  purchase_amount):
#  Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за
#  допомогою scatter plot.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)', color='skyblue')
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупок')
plt.show()
```

```python
#  7. Сезонність покупок (season):
#  Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти
#  роблять найбільше покупок.

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='season', order=df['season'].value_counts().index)
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#  8. Графік розподілу розмірів одягу (size):
#  Створіть графік розподілу розмірів одягу, щоб визначити популярність
#  різних розмірів.

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='size', palette='Set2', order=df['size'].value_counts().index)
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#  9. Кількість покупок відносно типу оплати (payment_method):
#  Порівняйте, скільки покупок робиться за різними методами оплати за
#  допомогою стовпчатої діаграми.

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='payment_method', palette='pastel', order=df['payment_method'].value_counts().index)
plt.title('Кількість покупок за методом оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45)
plt.show()
```

```python
#  10.Залежність між кількістю попередніх покупок (previous_purchases) і
#  сумою поточної покупки (purchase_amount):

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='previous_purchases', y='purchase_amount_(usd)', color='skyblue')
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума поточної покупки')
plt.show()
```

```python
# ЗАВДАННЯ 2
#
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності
# між трьома числовими змінними з використанням бібліотеки Seaborn.
# Налаштуйте відображення графіка, включаючи назву графіку, підписи
# осей і інші необхідні параметри для кращого розуміння візуалізації.
# Додайте кольорову легенду.
#  𝑧 = 𝑠𝑖𝑛 √𝑥
#  2 + 𝑦
#  2, 𝑥 ∈ [−5; 5], 𝑦 ∈ [−5; 5]

# Генерація даних
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Побудова графіка поверхні
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('Залежність sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```
